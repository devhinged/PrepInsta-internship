```python
make_counts = df['make'].value_counts()

# Plotting pie chart
plt.figure(figsize=(8,8))
plt.pie(make_counts, labels=make_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Car Makes')
plt.show()
```

**Pie Chart:**

The pie chart visualizes the distribution of car makes in the dataset.
Each slice of the pie represents a car make, with its size proportional to the number of occurrences of that make in the dataset.
The labels on the pie chart display the names of the car makes, and the percentage value represents the proportion of each make relative to the total.
By observing the pie chart, we can quickly understand the relative frequency or distribution of different car makes in the dataset.

```python
df['ex-showroom_price'] = df['ex-showroom_price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)

# Grouping by make and calculating average ex-showroom price
avg_price = df.groupby('make')['ex-showroom_price'].mean().reset_index()

# Plotting bar chart
plt.figure(figsize=(15, 6))
plt.bar(avg_price['make'], avg_price['ex-showroom_price'], color='skyblue')
plt.xlabel('Make')
plt.ylabel('Average Ex-showroom Price (INR)')
plt.title('Average Ex-showroom Price by Make')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Bar Chart:**

The bar chart visualizes the average ex-showroom price for each car make.
It allows for easy comparison of average prices across different makes.
The x-axis represents the car makes, while the y-axis represents the average ex-showroom price in INR (Indian Rupees).

```python
import seaborn as sns
plt.figure(figsize=(30, 6))
sns.barplot(data=df, x='make', y='ex-showroom_price', hue='fuel_type')
plt.title('Bar Plot with Hues: Ex-showroom Price by Make and Fuel Type')
plt.xlabel('Make')
plt.ylabel('Ex-showroom Price')
plt.legend(title='Fuel Type')
plt.show()

# Heatmap
plt.figure(figsize=(20, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Joint plot
sns.jointplot(data=df, x='seating_capacity', y='ex-showroom_price', kind='scatter')
plt.suptitle('Joint Plot: Seating Capacity vs Ex-showroom Price', y=1.02)
plt.show()
```

**Bar Plot with Hues (Ex-showroom Price by Make and Fuel Type):**

The bar plot with hues displays the distribution of ex-showroom prices across different car makes, distinguished by fuel types.
It helps in comparing the average ex-showroom prices for each make across different fuel types.
The legend shows the fuel types represented by different colors, aiding in interpretation.

**Correlation Heatmap:** **bold text**

The heatmap visualizes the correlation coefficients between numerical variables in the dataset.
It provides insights into the strength and direction of relationships between variables.
Darker shades indicate stronger correlations, while lighter shades indicate weaker correlations or no correlation.

**Joint Plot (Seating Capacity vs Ex-showroom Price):**

The joint plot illustrates the relationship between seating capacity and ex-showroom prices.
It includes both a scatter plot of the data points and histograms showing the distributions of each variable.
Helps in understanding the correlation and distribution of these variables simultaneously.

```python
correlation_matrix = df.corr()

# Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='ex-showroom_price', y='seating_capacity')
plt.title('Scatter Plot: Ex-showroom Price vs Seating Capacity')
plt.xlabel('Ex-showroom Price')
plt.ylabel('Seating Capacity')
plt.show()

# Pair plot
sns.pairplot(df[['ex-showroom_price', 'seating_capacity', 'cylinders']])
plt.suptitle('Pair Plot', y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

**Scatter Plot (Ex-showroom Price vs Seating Capacity):**

The scatter plot illustrates the relationship between ex-showroom prices and seating capacity.
There doesn't seem to be a strong linear relationship between ex-showroom prices and seating capacity based on the scatter plot.

**Pair Plot:**

The pair plot displays pairwise relationships between ex-showroom price, seating capacity, and cylinders.
It allows for a quick overview of the correlations and distributions between these variables.

**Correlation Matrix and Heatmap:**

The correlation matrix and heatmap visualize the correlation coefficients between variables.
It provides insights into the strength and direction of relationships between ex-showroom price, seating capacity, and cylinders.
The heatmap makes it easier to identify strong and weak correlations, with values ranging from -1 to 1.

```python
plt.figure(figsize=(10, 6))
sns.catplot(data=df, x='fuel_type', y='ex-showroom_price', kind='bar', hue='make', palette='viridis')
plt.title('Categorical Plot: Ex-showroom Price by Fuel Type and Make')
plt.xlabel('Fuel Type')
plt.ylabel('Ex-showroom Price')
plt.show()

# Stacked bar chart
df_grouped = df.groupby('make')['fuel_type'].value_counts().unstack().fillna(0)
df_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Chart: Fuel Type Distribution by Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.legend(title='Fuel Type')
plt.show()

# Violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='fuel_type', y='ex-showroom_price', hue='make', split=True, palette='muted')
plt.title('Violin Plot: Ex-showroom Price by Fuel Type and Make')
plt.xlabel('Fuel Type')
plt.ylabel('Ex-showroom Price')
plt.show()
```

**Categorical Plot (Ex-showroom Price by Fuel Type and Make):**

Shows ex-showroom prices distribution across fuel types and car makes.
Highlights significant price variations among different fuel types and car makes.
Tata cars fueled by CNG tend to have lower ex-showroom prices compared to petrol-fueled ones.
Within each fuel type, there are price variations suggesting other factors influencing prices.

**Stacked Bar Chart (Fuel Type Distribution by Make):**
**bold text**
Illustrates the distribution of fuel types among different car makes.
Each bar represents a car make, and segments represent the distribution of fuel types.
Most Tata cars are fueled by petrol, followed by CNG, with minimal representation of other fuel types like diesel or electric.

**Violin Plot (Ex-showroom Price by Fuel Type and Make):**

Displays the distribution of ex-showroom prices for each fuel type and car make.
Allows visual comparison of price distributions across fuel types and makes.
Indicates price variations among different fuel types and makes, with petrol-fueled Tata cars having a wider price range compared to CNG-fueled ones.
