```python
import numpy as np
import pandas as pd
import matplotlib as plt
data = pd.read_csv("dev.csv")
data
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.tail()
```

```python
x= data.columns
x
```

```python
avg_age = data['Age'].mean()
data['Age']= data['Age'].fillna(avg_age)

#descriptive analysis
data['Age']
```

```python
data['Salary']=data['Salary'].apply((lambda x: x.replace('$',"")))
data['Salary']
```

```python
data = data.drop(['Index'],axis=1)
```

```python
data
```

```python
neg=data[data['Established']<0]
neg
```

```python
med = data['Established'].median()
med
```

```python
data.loc[data['Established']<0,'Established']=med
data
```

```python
data.loc[data['Easy Apply']=='TRUE','Easy Apply']=1
```

```python
data['Easy Apply']=data['Easy Apply'].astype(int)
data.loc[data['Easy Apply']<0,'Easy Apply']=0

data
```

```python
avg_rating = data['Rating'].mean()
data['Rating']= data['Rating'].fillna(avg_rating)
data['Rating']
```

```python
data
```

```python
import seaborn as sns


sns.set(style="whitegrid")

fig, axes = plt.subplots(nrows=1, ncols=len(data.columns), figsize=(20, 6))
for i, column in enumerate(data.columns):
    sns.boxplot(x=data[column], ax=axes[i])
    axes[i].set_title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```
